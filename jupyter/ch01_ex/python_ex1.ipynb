```python
import pandas as pd
df = pd.DataFrame({'A' : ['데이터 분석', '기본 학습서', '퇴근 후 열공'],
                   'B' : [10, 20, 30],
                   'C' : ['ab cd', 'AB CD', 'ab cd'] })
# 문자열 변경
df['A'] = df['A'].str.replace('분석', '시각화')
# 숫자 변경
df['B'] = df['B'].replace(10, 100)
# 문자열 분리
df['A'].str.split()
# 특정 행의 문자열 분리
df['A'].str.split()[0]
# D라는 열 추가후 A열의 문자열을 분리한 뒤 첫번째 단어를 삽입
df['D'] = df['A'].str.split().str[0]
# 특정 문자열 검색
df['A'].str.contains('기본')
# A열의 문자열 중 기본이 포함된 문자열의 유무를 기본포함유무 추가 후 삽입
df['기본포함유무'] = df['A'].str.contains('기본')
# 문자열 길이
df['문자길이'] = df['A'].str.len()
# 대소문자 변경
df['C'] = df['C'].str.lower()
df['C'] = df['C'].str.upper()
# 공백 처리
df['C'] = df['C'].str.lower()
df['C'] = df['C'].str.replace(" ", "")
# 문자열 슬라이싱
df['C'].str[1:3]
df['C'][1:3]
```

```python
# date_data의 DataFrame 중 DateTime4 열만을 불러올 수 있음, parse_dates를 사용하면 날짜로 자동변환
df = pd.read_csv('date_data.csv', usecols = ['DateTime4'], parse_dates = ['DateTime4'])
# 특정 날짜로부터 100일 기념일 계산
day = pd.Timedelta(days=99)
df['100day'] = df['DateTime4'] + day
# 특정 날짜로부터 100시간 뒤
hour = pd.Timedelta(hours=100)
df['100hour'] = df['DateTime4'] + hour
# 특정 날짜로부터 7주, 7일, 7시간, 7분, 7초 이전
difference = pd.Timedelta(weeks=7, days=7, hours=7, minutes=7, seconds=7)
df['difference'] = df['DateTime4'] - difference
# 두 시간 사이의 차이
diff = df['100hour'] - df['difference']
print(diff.dt.total_seconds())
print(diff.dt.total_seconds()/60)
print(diff.dt.total_seconds()/60/60)
print(diff.dt.total_seconds()/60/60/24)
print(diff.dt.days)
print(diff.dt.seconds)
# 시간 반올림
min = 5.41
print(int(min), "분")
print(0.41*60, "ch")
print(round(diff.dt.total_seconds()/60))
```
